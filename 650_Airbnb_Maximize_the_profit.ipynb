```python
import pandas as pd
f = pd.read_csv("C:/Users/Administrator/Desktop/listings.csv")
f.head()
y_data = f["price"]
y_data = pd.DataFrame(y_data)
y_data.head()
```

```python
x_data = f.select_dtypes(include=['float64','int64'])
x_data = pd.DataFrame(x_data)
x_data.head()
```

```python
import numpy as np
my_data = x_data.join(y_data)
my_data.head()
my_data.columns
```

```python
my_data.to_csv("data.csv", index=False)
```

```python
corr_matx = my_data.corr()
corr_matx
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.heatmap(corr_matx)
plt.show()
```
