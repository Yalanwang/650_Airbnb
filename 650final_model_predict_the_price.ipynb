```python
train_1 = train_data1.dropna()
train_2 = train_data2.dropna()
train_3 = train_data3.dropna()
```

```python
test_1 = test_data1.dropna()
test_2 = test_data2.dropna()
test_3 = test_data3.dropna()
```

```python
train1 = train_1.select_dtypes(include=['float64','int64']).iloc[:,1:]
train2 = train_2.select_dtypes(include=['float64','int64']).iloc[:,1:]
train3 = train_3.select_dtypes(include=['float64','int64']).iloc[:,1:]
```

```python
x_test_1 = test_1.select_dtypes(include=['float64','int64']).iloc[:,1:]
x_test_2 = test_2.select_dtypes(include=['float64','int64']).iloc[:,1:]
x_test_3 = test_3.select_dtypes(include=['float64','int64']).iloc[:,1:]
```

```python
y_train_1 = train1["price"]
x_train_1 = train1.drop("price",axis = 1)

y_train_2 = train2["price"]
x_train_2 = train2.drop("price",axis = 1)

y_train_3 = train3["price"]
x_train_3 = train3.drop("price",axis = 1)
```

```python
x_train_1.head()
```

```python
x_train_2.head()
```

```python
x_train_3.head()
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train_1, y_train_1)
probs1 = model1.predict(x_test_1)
y_test_1 = pd.DataFrame(probs1)
y_test_1.head()
```

```python
model2 = LogisticRegression()
model2.fit(x_train_2, y_train_2)
probs2 = model2.predict(x_test_2)
y_test_2 = pd.DataFrame(probs2)
y_test_2.head()
```

```python
model3 = LogisticRegression()
model3.fit(x_train_3, y_train_3)
probs3 = model3.predict(x_test_3)
y_test_3 = pd.DataFrame(probs3)
y_test_3.head()
```

```python
out_df1 = pd.DataFrame(y_test_1)
out_df1.columns = ["price"]

out_df1.to_csv("650_pre_1.csv", index=False)
```
